```python
left = pd.DataFrame({"key1": ["foo", "foo", "bar"],
                     "key2": ["one", "two", "one"],
                     "lval": pd.Series([1, 2, 3],dtype="Int64")})
```

```python
right = pd.DataFrame({"key1": ["foo", "foo", "bar", "bar"],
                     "key2": ["one",  "one", "one", "two",],
                     "lval": pd.Series([4, 5, 6, 7],dtype="Int64")})
```

```python
pd.merge(left, right, on=["key1", "key2"], how="outer")
```

```python
pd.merge(left, right, on="key1")
```

```python
lefth = pd.DataFrame({"key1":["Ohio", "Ohio", "Ohio",
                              "Nevaba", "Nevaba"],
                      "key2":[2000, 2001, 2002, 2001, 2002],
                      "data": pd.Series(range(5), dtype="Int64")})
```

```python
right_index = pd.MultiIndex.from_arrays(
    [
        ["Nevaba", "Nevaba", "Ohio", "Ohio", "Ohio", "Ohio"],
        [2001, 2000, 2000, 2000, 2001, 2002]
    ]
)
```

```python
right = pd.DataFrame({"event1": pd.Series([0, 2, 4, 6, 8, 10], dtype="Int64",
                                          index=right_index),
                      "event2": pd.Series([1, 3, 5, 7, 9, 11], dtype="Int64",
                                          index=right_index)})
```

```python
lefth
```

```python
right
```

```python
pd.merge(lefth, right, left_on=["key1", "key2"], right_index=True)
```
