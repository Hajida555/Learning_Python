```python
def drawlight():
    if r.randint(0, 30) == 0:
        color('tomato')
        circle(6)
    elif r.randint(0, 30) == 1:
        color('orange')
        circle(3)
    else:
        color('dark green')
```

```python
backward(n*4.8)
```

```python
def tree(d, s):
    if d <= 0: return
    forward(s)
    tree(d-1, s*.8)
    right(120)
    tree(d-3, s*.5)
    drawlight()
    right(120)
    tree(d-3, s*.5)
    right(120)
    backward(s)
```

```python
tree(15, n)
```

```python
def drawsnow():
    t.ht()
    t.pensize(2)
    for i in range(200):
        t.pencolor('white')
        t.pu()
        t.setx(r.randint(-450,450))
        t.sety(r.randint(-150,450))
        t.pd()
        dens = 6
        snowsize = r.randint(1,10)
        for i in range(dens):
            #t.forward(int(snowsize))
            t.fd(int(snowsize))
            t.backward(int(snowsize))
            #t.bd(int(snowsize))
            t.right(int(360/dens))
```

```python
drawsnow()
```
